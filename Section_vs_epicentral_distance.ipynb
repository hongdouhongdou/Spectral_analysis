```python

# Now read all, process, cut and plot section
tp_manual=[]
ts_manual=[]
for test in fnames:
    fname_Z=test[3]+test[4]+test[2]
    st_tmp=read(fname_Z)
    # Get origin time
    sac=SacIO(fname_Z)
    #sac.ListStdValues()
    to=sac.GetHvalue('o')
    tb=sac.GetHvalue('b')
    tp=sac.GetHvalue('a')
    ts=sac.GetHvalue('T0')
    if tp == -12345 or ts == -12345:
        print 'tp or ts manual undefined in file '+fname_Z
        tp=sac.GetHvalue('T1')
        ts=sac.GetHvalue('T2')
    nzyear=sac.GetHvalue('nzyear')
    nzjday=sac.GetHvalue('nzjday')
    nzhour=sac.GetHvalue('nzhour')
    nzmin=sac.GetHvalue('nzmin')
    nzsec=sac.GetHvalue('nzsec')
    nzmsec=sac.GetHvalue('nzmsec')
    t1=UTCDateTime(str(nzyear)+"-"+str(nzjday-1).zfill(3))
    t_ref=UTCDateTime(str(nzyear)+"-"+str(nzjday).zfill(3)+"T"+str(nzhour).zfill(2)+":"+str(nzmin).zfill(2)+":"+str(nzsec).zfill(2)+"."+
                  str(nzmsec).zfill(2))
    t_origin=t_ref+to-tb
    tp_manual.append(tp-to+tb)
    ts_manual.append(ts-to+tb)
    st_tmp.cutout(t1,t_origin)
    try:
        st
        st+=st_tmp
    except:
        st=st_tmp
st.normalize()
st.detrend()
st.taper(max_percentage=0.05, type='cosine')
st.filter('bandpass', freqmin=0.5, freqmax=20, corners=2, zerophase=True)

#print st[0].times()
#print len(st[0].data)
print len(st),len(tp_manual),len(fnames)
# plot type section needs tr.stats.distance
for tr in st:
    tr.stats.distance=tr.stats.sac.dist*1000
    #print tr.times()
```

```python

dado={}
dado['time']=[i.times() for i in st]
dado['amp']=[i.data for i in st]
dado['R']=[sqrt(pow(i.stats.sac['dist'],2)+pow(i.stats.sac['evdp'],2)) for i in st]
dado['tp_manual']=tp_manual
dado['ts_manual']=ts_manual
dado['orglist'] = np.argsort([i for i in dado['R']])

print len(dado['tp_manual']),len(dado['R']),len(dado['tp_manual'])
```

## Create a polynomial order 2 tp model

```python
weights=np.ones_like(dado['R'])
for k, tp in enumerate(dado['tp_manual']):
    # Give zero weight to wrong tp (<0 or > 200 sec) or R >1000km
    if (tp <= 0 or tp > 200 or dado['R'][k] > 1000) or (tp > 40 and dado['R'][k]< 150) or (tp > 60 and dado['R'][k]< 300) or (tp <20 and dado['R'][k]> 300) or (tp > 120 and dado['R'][k]< 800):
        weights[k]=0.0
p=np.polyfit(dado['R'],dado['tp_manual'],2,w=weights)
print p
tp_model = np.poly1d(p)

plt.scatter(dado['R'],dado['tp_manual'],marker='+',color='red')
plt.plot(np.arange(0,1000,10),tp_model(np.arange(0,1000,10)),color='blue')
plt.xlim(xmin=-10,xmax=1000)
plt.ylim(ymin=-10,ymax=200)
```

## Create a polynomial order 2 ts model

```python
del weights,p
weights=np.ones_like(dado['R'])
Rdata=[];
Tdata=[];
for k, ts in enumerate(dado['ts_manual']):
    # Give zero weight to wrong tp (<0 or > 200 sec) or R >1000km
    if (ts <= 0 or ts > 400 or dado['R'][k] > 1000) or (ts > 30 and dado['R'][k]< 100) or (ts > 60 and dado['R'][k]< 200) or (ts < 50 and dado['R'][k]> 200) or (ts > 80 and dado['R'][k]< 300) or (ts >175 and dado['R'][k]< 700) or (ts < 150 and dado['R'][k]> 800):
        weights[k]=0.0
        Rdata.append(dado['R'][k])
        Tdata.append(ts)
p=np.polyfit(dado['R'],dado['ts_manual'],2,w=weights)
print p
ts_model = np.poly1d(p)

plt.scatter(dado['R'],dado['ts_manual'],marker='+',color='red')
plt.scatter(Rdata,Tdata,marker='+',color='blue')
plt.plot(np.arange(0,1000,10),ts_model(np.arange(0,1000,10)),color='blue')
plt.xlim(xmin=-10,xmax=1000)
plt.ylim(ymin=-10,ymax=400)
```
