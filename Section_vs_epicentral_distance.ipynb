## Create a polynomial order 2 tp model

```python
weights=np.ones_like(dado['R'])
Rdata=[];
Tdata=[];
for k, tp in enumerate(dado['tp_manual']):
    # Give zero weight to wrong tp (<0 or > 200 sec) or R >1000km
    if (tp <= 0 or tp > 200 or dado['R'][k] > 1000) or (tp > 40 and dado['R'][k]< 150) or (tp > 60 and dado['R'][k]< 300) or (tp <20 and dado['R'][k]> 300) or (tp > 120 and dado['R'][k]< 800):
        weights[k]=0.0
        Rdata.append(dado['R'][k])
        Tdata.append(tp)
p=np.polyfit(dado['R'],dado['tp_manual'],2,w=weights)
tp_model_ori = np.poly1d(p)
# Set the constant term to 0.0
tp_model = np.poly1d([p[0],p[1],0.0])

print 'tp model: tp= {0:.3e} *Rhypo^2 + {1:.3e} *Rhypo'.format(p[0],p[1])

plt.figure(figsize=(10,6))
red_crosses = plt.scatter(dado['R'],dado['tp_manual'],marker='+',color='red',label='Manual tp picks')
black_crosses = plt.scatter(Rdata,Tdata,marker='+',color='black',label='data discarded for fit')
blue_line, = plt.plot(np.arange(0,1000,10),tp_model(np.arange(0,1000,10)),color='blue',label='polynomial order 2 modified (cste=0.0)')
blue_dashed_line, = plt.plot(np.arange(0,1000,10),tp_model_ori(np.arange(0,1000,10)),color='blue',linestyle='--',label='polynomial order 2 original')
plt.xlim(xmin=-10,xmax=1000)
plt.ylim(ymin=-10,ymax=200)
plt.legend((red_crosses, black_crosses, blue_line, blue_dashed_line),
           ('Manual tp picks', 'data discarded for fit', 'polynomial order 2 modified (cste=0.0)', 'polynomial order 2 original'))
plt.xlabel('Hypocentral distance (km)',fontsize=14)
plt.ylabel('Time (s)',fontsize=14)

plt.savefig('model_tp.png')
```

```python
# Define the fit function
def quadratic_fit_with_fixed_origin(x, y):
 '''Fits a quadratic fit of the form ax^2+bx to the data (i.e. y(x=0)=0)'''
 fitfunc = lambda params, x: params[0] * pow(x,2) + params[1] * x #create fitting function of form mx+b
 errfunc = lambda p, x, y: fitfunc(p, x) - y #create error function for least squares fit

 init_a = 0.1 #find initial value for a (gradient)
 init_b = 1 #find initial value for b (y axis intersection)
 init_p = numpy.array((init_a, init_b)) #bundle initial values in initial parameters

 #calculate best fitting parameters (i.e. m and b) using the error function
 p1, success = scipy.optimize.leastsq(errfunc, init_p.copy(), args = (x, y))
 f = fitfunc(p1, x) #create a fit with those parameters
 return p1, f 
```
